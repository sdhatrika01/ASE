### Import Libraries

```python
import cv2
import pandas as pd
import numpy as np
from glob import glob
import hashlib
import os
from tensorflow.keras.preprocessing.image import load_img,img_to_array
from tensorflow.keras.models import model_from_json
from tensorflow.keras import Model
from pathlib import Path
```

```python
import os
file_path=[]
files=os.listdir("data")
for i in range(len(files)):
    file_path.append(os.path.join("data",files[i]))
```

```python
print(file_path)
```

```python
count = 0

for i in range(0,6):
    cam = cv2.VideoCapture(file_path[i])
    currentframe = 0
    while(True): 
        ret,frame = cam.read() 
        if ret:
            if count%1 == 0 :
                filename = 'Extracted_Frames/'+file_path[i].split('\\')[1]+"_frame%d.jpg" % count;
                print ('Creating...' + filename) 
                cv2.imwrite(filename, frame) 
                currentframe += 1
            count+=1
        else: 
            break
    cam.release() 
    #cv2.destroyAllWindows() 
```

```python
file = Path("cnn_model/model_structure.json")
model_structure = file.read_text()
model = model_from_json(model_structure)
model.load_weights("cnn_model/model_weights.h5")
```

```python
for idx in range(len(model.layers)):
    print(model.get_layer(index = idx).name)
```

```python
intermediate_layer_model = Model(inputs=model.input,outputs=model.get_layer('dense').output)
```

```python
images = glob("Extracted_Frames/*.jpg")
print("Number of frames:{}".format(len(images)))
```

```python
print(images)
```

```python
train_image = []
frame_names = []
video_name = []
for i in range(len(images)):
    img = load_img(images[i], target_size=(64,64,3))
    img = img_to_array(img)
    img = img/255.0
    train_image.append(img)
    f_name=images[i]
    f_names=f_name.split("\\")[1]
    f_names=f_names.split(".jpg")[0]
    f_names=f_names.split(".mp4_")[1]
    frame_names.append(f_names)
    f1_names=f_name.split("\\")[1]
    f1_names=f1_names.split(".mp4_")[0]
    video_name.append(f1_names)
X = np.array(train_image)
```

```python
intermediate_output = intermediate_layer_model.predict(X)
```

```python
df=pd.DataFrame(intermediate_output)
```

```python
df.shape
```

```python
column_names=[]
for i in range(1024):
    column_names.append("fet_"+str(i))
```

```python
df.columns=column_names
```

```python
df.insert(loc=0,column="VideosNames",value=video_name)
```

```python
df.insert(loc=1,column="FramesNames",value=frame_names)
```

```python
df.head()
```

```python
df.to_csv("Feature_Files/LocalFeatureExtraction.csv",index=False)
```
